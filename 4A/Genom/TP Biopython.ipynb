## Exercice 2 : Analyse de séquences

```python
from Bio import SeqUtils

records = list(SeqIO.parse("seq.fasta", "fasta"))
print(records[0].id)  # first record
print(records[-1].id)  # last record
print(records)
print(records[0].seq)
print(SeqUtils.GC(records[0].seq))


with open('resume.txt','w') as f:
    f.write('ID' + '\t|\t' + 'Sequence length' + '\t|\t' + 'GC%' + '\n')
    for i in range(len(records)):
        f.write(records[i].id + '\t|\t' + str(len(records[i].seq)) + '\t|\t' +str(round(SeqUtils.GC(records[i].seq), 2)) + '\n')
        
        
```

## Exercice 3 : Complément inverse et traduction

```python
records[0].seq.reverse_complement()
print(list(SeqIO.parse('sequences.gdb', "genbank")))
```

```python
def new_prog(file):
    records = list(SeqIO.parse(file, "genbank"))
    comp_records = records
    for i in range(len(records)):
        comp_records[i].seq = records[i].seq.reverse_complement()
        
    
    with open('reverse_complementary.fasta','w') as f:
        for i in range(len(comp_records)):
            SeqIO.write(comp_records[i],f, "fasta")
            
    trans_records = records
    for i in range(len(records)):
            trans_records[i].seq = records[i].seq.translate()
            
    with open('translated.fasta','w') as f:
        for i in range(len(trans_records)):
            SeqIO.write(trans_records[i],f, "fasta")
            
            
new_prog('sequences.gdb')            
```
